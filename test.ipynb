```python
#Load and diaplay image
image = cv2.imread('data/house1.png')
image = colors.bgr2rgb(image)
plt.imshow(image)
```

```python
fig, ax = plt.subplots(4, 3) # rows: RGB, HSV, XYZ, LAB and cols: OpenCV, Custom, Reprojection
fig.suptitle('Color Space Tests')

#display built int for comparisons
xyzcv = cv2.cvtColor(image, cv2.COLOR_RGB2XYZ)
ax[0,0].set_xlabel('OpenCV XYZ')
ax[0,0].imshow(xyzcv)

labcv = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
ax[0,1].set_xlabel('OpenCV Lab')
ax[0,1].imshow(labcv)

#Convert to xyz and test
xyz = colors.rgb2xyz(image)
xyz_msg = 'XYZ color space'
ax[1,0].set_xlabel(xyz_msg)
ax[1,0].imshow(xyz)

#Convert to Lab and test
lab = colors.rgb2lab(image)
lab_msg = 'Lab color space'
ax[1,1].set_xlabel(lab_msg)
ax[1,1].imshow(lab)

#Convert to rgb and test
rgb1 = colors.xyz2rgb(xyz)
xyz_msg = 'XYZ to RGB'
ax[2,0].set_xlabel(xyz_msg)
ax[2,0].imshow(rgb1)

#Convert to Lab and test
rgb2 = colors.lab2rgb(image)
lab_msg = 'Lab ro RGB'
ax[2,1].set_xlabel(lab_msg)
ax[2,1].imshow(colors.xyz2rgb(colors.rgb2xyz(image)))#rgb1)
```
